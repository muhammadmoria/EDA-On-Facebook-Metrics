```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("Facebook Metrics of Cosmetic Brand.csv")
```

```python
df.head()
```

```python
df = df.drop('Unnamed: 0',axis=1)
```

```python
df
```

```python
print(df.info())
print('----------------------------------')
print(df.describe().T)
```

Total likes ditribution

```python
"""----------------------------------------------Total likes ditribution-------------------------------------------------------"""
sns.histplot(df['Page total likes'], kde=True)
plt.show()
```

```python
"""-------------------------------------------------Count of Post Types-----------------------------------------------------------------"""
sns.countplot(x='Type', data=df)
plt.show()
```

```python
"""-------------------------------------------------Paid vs Unpaid Posts-----------------------------------------------------------"""
sns.countplot(x='Paid', data=df)
plt.show()
```

```python
"""-------------------------------------------------Lifetime Post Total Reach-----------------------------------------------------------"""
sns.histplot(df['Lifetime Post Total Reach'], kde=True)
plt.show()
```

```python
"""-------------------------------------------------Post Type VS Total Interactions-----------------------------------------------------------"""
sns.boxplot(x='Type', y='Total Interactions', data=df)
plt.show()
```

```python
"""-------------------------------------------------Paid vs Lifetime Post Total Reach-----------------------------------------------------------"""
sns.boxplot(x='Paid', y='Lifetime Post Total Reach', data=df)
plt.show()
```

```python
"""-------------------------------------------------Post Hour vs Total Interactions-----------------------------------------------------------"""
sns.lineplot(x='Post Hour', y='Total Interactions', data=df)
plt.show()
```

```python
"""-------------------------------------------------Total Interactions vs Post Weekday-----------------------------------------------------------"""
plt.figure(figsize=(12, 6))
sns.lineplot(x='Post Weekday', y='Total Interactions', data=df, marker='o')
plt.xticks(range(1, 8), ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.show()
```

```python
"""-------------------------------------------------Engagement Rate-----------------------------------------------------------"""
df['Engagement Rate'] = df['Total Interactions'] / df['Lifetime Post Total Reach']
plt.figure(figsize=(12, 6))
sns.histplot(df['Engagement Rate'], kde=True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Lifetime Post Total Reach', y='Lifetime Engaged Users', data=df)
plt.show()
```

```python
metrics = ['comment', 'like', 'share', 'Total Interactions']
for metric in metrics:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x='Type', y=metric, data=df)
    plt.title(f'{metric.capitalize()} by Post Type')
    plt.xlabel('Post Type')
    plt.ylabel(metric.capitalize())
    plt.show()
```

```python
df_encoded = pd.get_dummies(df, columns=['Type'], drop_first=True)
corr_matrix = df_encoded.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.show()
```
